```python
def DogruBolgeleriBul():
    for i in df.index: #i veri setinin girdileri içerisinde tek tek gezer
        bolge=df.iloc[i]["Region"] #her girdinin bölgelerini tutar bolge degiskeni
        if str(bolge)=="nan": #bu bölge degiskeni nan mi diye kontrol edilir
                                #eger nan ise
            olasıBolgeler=df[(df["Country"]==df.iloc[i]["Country"]) & df["Region"].notnull()]
                                     #veri tabanında bölgesi nan olan o girdiye ait baska ülkeler var mı
                                         #ve bu girdilerden de bölgeleri nan olamayan girdileri alınır
            
            if olasıBolgeler.shape[0] > 0: #olası bölgelerin girdi sayısı 0 ise 
                                            #yani diger bir deyişle bölgesi nan olan ülkenin baska bir girdilerde nan olmayan
                                            #bölgesi var ise
                dogruBolge=olasıBolgeler.iloc[0]["Region"]
                #ilk girdinin bölge degeri alınır
                
            else:  #aksi takdirde o bölgenin degeri oldugu gibi tekrar alınır yani nan olsa bile
                dogruBolge=bolge
            df.loc[i,"Region"]=dogruBolge #ve o girdinin bölge degeri yeniden güncellenir
```
