```python
colrGradient = {0.0: 'blue',
                0.20: 'cyan',
                0.40: 'lime',
                0.60: 'yellow',
                0.80:'orange',
                1.0: 'red'}
```

```python
def HeatMapByColumnAndYear(column,year):
    map_hooray = folium.Map(location=[51.5074, 0.1278],
                    zoom_start = 5,tiles="Cartodb Positron")
    df['latitude'] = df['latitude'].astype(float)
    df['longitude'] = df['longitude'].astype(float)
    heat_df=df[df['Year']==year]
    heat_df = heat_df[['latitude', 'longitude',column]]
    heat_data = [[row['latitude'],row['longitude'],row[column]] for index, row in heat_df.iterrows()]
    HeatMap(heat_data, gradient=colrGradient).add_to(map_hooray)
    display(map_hooray)
```

```python
HeatMapByColumnAndYear("Happiness Score",2019)
```
